```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
path = './sample/dog.jpeg'

image_pil = Image.open(path)
image = np.array(image_pil)
```

```python
#이미지 정보 확인
print(image.shape)

#image range 확인
print(np.min(image), np.max(image))

#image visualization
plt.hist(image.ravel(),256,[0,256])
plt.show()
```

```python
#이미지 보기
plt.imshow(image)
plt.show()
```

```python
#이미지 흑백으로 열기
image_pil = Image.open(path).convert("L")
image_bw = np.array(image_pil)

plt.imshow(image_bw,'gray')
plt.show()
```

```python
#이미지 red, blue로 열기
plt.imshow(image_bw,'RdBu')
plt.show()
```

```python
#이미지 jet로 열기
plt.imshow(image_bw, 'jet')
plt.show()
```

```python
#color bar add
plt.imshow(image_bw, 'jet')
plt.colorbar()
plt.show()
```

```python
#image resize
!pip install opencv-python
import cv2

dog_image = cv2.resize(image,(275,183))
dog_image.shape
```

```python
path_cat='./sample/cat.jpeg'
image_pil = Image.open(path_cat)
cat_image = cv2.resize(np.array(image_pil),(275,183))


#두개의 이미지 겹치기
plt.imshow(dog_image)
#alpha:투명도
plt.imshow(cat_image,alpha = 0.5)
plt.show()
```

```python
#여러개의 이미지 출력
plt.figure(figsize=(10,10))

plt.subplot(221)
plt.imshow(dog_image)

plt.subplot(222)
plt.imshow(image_bw,'gray')

plt.subplot(223)
plt.imshow(cat_image)

plt.show()
```
